```python
#Importing essential libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Link to the dataset: https://www.kaggle.com/datasets/imtkaggleteam/breast-cancer/data

```python
#Loading the dataset
df = pd.read_csv('Data/breast-cancer-wisconsin-data_data.csv')
df.head()
```

## Exploring the data

```python
print(f'Number of rows: {len(df)}')
print(f'Number of columns: {len(df.columns)}')
```

```python
df.dtypes
```

```python
#Checking missing values
df.isna().sum()
```

```python
#Since we have only 1 object column, let's see whether that's the label of this dataset
df['diagnosis'].value_counts()
```

The column `Diagnosis` is the label (class) in this dataset. `B- benign` & `M- malignant`

Furthermore, we can drop `Unnamed: 32` & `id` colums since they won't contribute in any way for classification

```python
df.drop(['id', 'Unnamed: 32'], axis=1, inplace= True)
df.columns
```

## Feature selection

```python
#Splitting labels and features
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

X.shape, y.shape
```

```python
#Training and testing splits
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3)
```

## Training a model

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train) #Fitting the data
```

The model is trained now. Let's retreive the predictions for `X_test` data and compare them with the actual values

```python
y_preds = model.predict(X_test)
y_preds.shape
```

## Evaluating Model Performance

```python
from sklearn.metrics import accuracy_score

model_accuracy = accuracy_score(y_test, y_preds)
print(f'Model shown to have a {round(model_accuracy,2)*100}% of accuracy')
```

```python
#Let's draw a confusion matrix and explore even more 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_preds)
cm
```

```python
#Visualizing the confusion matrix
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels= model.classes_)
display.plot()
```

```python
#Let's evaluate furthermore, using precision, recall and f1 score as well
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_preds, pos_label= 'B') #setting a default positive label in binary classification is required
recall = recall_score(y_test, y_preds, pos_label= 'B')
f1 = f1_score(y_test, y_preds, pos_label= 'B')
```

```python
#Printing the values
print(f'Precision Score: {round(precision,2)*100}%')
print(f'Recall Score: {round(recall,2)*100}%')
print(f'F1 Score: {round(f1,2)*100}%')
```

A little note about what the above metrics mean:

* `Precision`: Measures the accuracy of positive predictions (i.e., the proportion of true positives among all predicted positives).
* `Recall`: Measures the model's ability to capture all positive instances (i.e., the proportion of true positives among all actual positives).
* `F1 Score`: The harmonic mean of precision and recall, useful when you want to balance precision and recall.
